```python
from random import shuffle

from insertion_sort import insertion_sort
from merge_sort import merge_sort, merge_sort_old
from radix_sort import radix_sort
from quicksort import quicksort, quicksort_plus_insertion_sort
```

```python
for i in range(1, 13):
    n = 2**i
    l = [j+1 for j in range(n)]
    shuffle(l)
    print(f"Merge Sort plus Insertion Sort for {n} items:")
    %timeit merge_sort(l[:])
    print()
    print(f"Radix Sort for {n} items:")
    %timeit radix_sort(l[:])
    print()
    print(f"Merge Sort for {n} items:")
    %timeit merge_sort_old(l[:])
    print()
    print(f"Quicksort plus Insertion Sort for {n} items:")
    %timeit quicksort_plus_insertion_sort(l[:])
    print()
    print(f"Quicksort for {n} items:")
    %timeit quicksort(l[:])
    print()
    print(f"Insertion Sort for {n} items:")
    %timeit insertion_sort(l[:])
    print("\n")

print("Done!")
```
